```python
import eqnet_format
```

```python
from eqnet_format import get_random_ksat, cnf_to_eqnet, ENCODER, dist_to_false, DPLL, mean, stdev
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy
```

```python
def gen_and_print(num, k, n, m):
    trues = []
    falses = []
    for i in range(num):
        rsat = get_random_ksat(k, n, m)
        clauses = [list(clause) for clause in rsat.clauses]
        eqnet_form = cnf_to_eqnet(clauses)
        encoding = ENCODER.get_encoding(eqnet_form)
        dist = dist_to_false(encoding)
        if DPLL().run(rsat):
            trues.append(dist)
        else:
            falses.append(dist)
    print("#Trues: {}; avg dist: {}; stdev dist: {}".format(
        len(trues), mean(trues), stdev(trues)))
    print("#Falses: {}; avg dist: {}; stdev dist: {}".format(
        len(falses), mean(falses), stdev(falses)))
    name = "{}-sats with {} vars and {} clauses".format(k, n, m)
    plt.title("Dist to false of True {}".format(name))
    plt.hist(trues, bins=numpy.linspace(0, 1.5, 21))
    plt.show()
    plt.title("Dist to false of False {}".format(name))
    plt.hist(falses, bins=numpy.linspace(0, 1.5, 21))
    plt.show()
```

```python
gen_and_print(1000, 2, 2, 3)
```
