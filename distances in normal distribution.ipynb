```python
from random import gauss
```

```python
from matplotlib import pyplot as plt
```

```python
S = 10000
STDEV = 1.
MEAN = 0.
```

```python
def dist(x, y):
    dist2 = sum((xe - ye)**2 for (xe, ye) in zip(x, y))
    dist = dist2 ** 0.5
    return dist
```

```python
def gen_example(dim):
    return tuple(gauss(MEAN, STDEV) for i in range(dim))
```

```python
def gen_arr(dim):
    return [gen_example(dim) for i in range(S)]
```

```python
arr1d = gen_arr(1)
```

```python
plt.hist(list(zip(*arr1d)), bins=100)
plt.show()
```

```python
for dist_from in [0.0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0, 2.5, 3.0]:
    dist1d = [dist(x, (dist_from,)) for x in arr1d]
    plt.title("Dist from {}".format(dist_from))
    plt.hist(dist1d, bins=100)
    plt.show()
```

```python
arr2d = gen_arr(2)
plt.scatter(*zip(*arr2d))
plt.show()
```

```python
for dist_from in [0.0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0, 2.5, 3.0]:
    dist2d = [dist(x, (dist_from, 0.0)) for x in arr2d]
    plt.title("Dist from {}".format(dist_from))
    plt.hist(dist2d, bins=100)
    plt.show()
```
